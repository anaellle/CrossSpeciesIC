```python
def construct_o2o_matrixanddf():
    # def the variables to return 
    matrix = []
    already_done = {}
    var_df = []
    cpt = 0
    tot = len(human_mouse_o2o.index)
    # start the loop throught the o2m dataframe
    for index, row in human_mouse_o2o.iterrows():
        #get the human id
        human_id = row.human_ensembl_gene_id
        
        # check if it has already been done, if not we get into the if
        if human_id not in already_done.keys():
            # starts new lines for dataframes
            var_line = {}
            
            # get all the human info
            var_line['human_ensembl_id'] = human_id
            var_line['human_entrez_id'] = row.human_entrezgene_id
            var_line['human_gene_name'] = row.human_external_gene_name
            var_line['orthology_type'] = row.orthology_type
            var_line['human_chromosome'] = homolog_human.var.loc[homolog_human.var.human_entrezgene_id == row.human_entrezgene_id].human_chromosome.iloc[0]
            var_line['human_long_gene_name'] = homolog_human.var.loc[homolog_human.var.human_entrezgene_id == row.human_entrezgene_id].human_gene_name.iloc[0]
            
            already_done[human_id] = []
            
            # get all the mouse homolog genes for the human gene
            current_df = human_mouse_o2o.loc[human_mouse_o2o.human_ensembl_gene_id == human_id]
            
            if len(current_df.mouse_homolog_ensembl_gene.values) == 1:
                var_line['mouse_ensembl_id'] = current_df.mouse_homolog_ensembl_gene.iloc[0]
                var_line['mouse_gene_name'] = current_df.mouse_homolog_gene_name.iloc[0]
                mouse_id = current_df.mouse_homolog_ensembl_gene.iloc[0]
            
            else :
                print('error')
            
            # add the newline to the count matrix
            if matrix == []:
                matrix = np.array([np.concatenate([homolog_human.X[:,np.where(homolog_human.var_names == homolog_human.var.loc[homolog_human.var.human_entrezgene_id == row.human_entrezgene_id].index[0])[0][0]],
                                          np.asarray(homolog_mouse.X[:,np.where(homolog_mouse.var_names == homolog_mouse.var.loc[homolog_mouse.var.index == human_id].index[0])[0][0]].todense().T)[0]])])
            else : 
                matrix = np.append(matrix,
                                   [np.concatenate([homolog_human.X[:,np.where(homolog_human.var_names == homolog_human.var.loc[homolog_human.var.human_entrezgene_id == row.human_entrezgene_id].index[0])[0][0]],
                                              np.asarray(homolog_mouse.X[:,np.where(homolog_mouse.var_names == homolog_mouse.var.loc[homolog_mouse.var.index == human_id].index[0])[0][0]].todense().T)[0]])],
                                   axis = 0)
            
            # To find the genes afterward if necessary
            already_done[human_id] = mouse_id

            
            # add the newline to the rows of the df
            var_df.append(var_line)
            
            cpt = cpt+1
            print(cpt,'/',tot,end='\r')
            
    return matrix.T, var_df, already_done
```
