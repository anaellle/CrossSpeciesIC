```python
import scanpy as sc
import pickle
import os
```

# SATURN Results for human and mouse

## Entire datasets

```python
adata = sc.read('saturn_results_1/test256_data_human_mouse_org_saturn_seed_0.h5ad')
adata
```

```python
sc.pp.pca(adata)
sc.pl.pca(adata, color="species", title="Species")
sc.pl.pca(adata, color="labels2", title="Cell Type") # The original cell type names
```

```python
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color="species", title="Species")
sc.pl.umap(adata, color="labels2", title="Cell Type")
```
